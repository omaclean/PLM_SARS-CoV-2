# Epistasis Benchmark

```python
innocenti = pd.read_csv('13059_2024_3355_MOESM1_ESM.csv')
innocenti = innocenti[['mi','outlier','gene_source','feature_codon_source','feature_codon_target','interaction']]
innocenti = innocenti[(innocenti.gene_source == 'S') & (innocenti.interaction == 'same gene')]
innocenti['feature_codon_source'] = innocenti.feature_codon_source.astype(int)
innocenti['feature_codon_target'] = innocenti.feature_codon_target.astype(int)
innocenti
```

```python
epi_data['source_position'] = epi_data.mutation.str[1:-1].astype(int)
```

```python
experimental = pd.read_csv('2ndorder_structure_summaryACE2.csv')
experimental['source'] = experimental.labels.str.split(',',expand=True)[0].astype(int)
experimental['target'] = experimental.labels.str.split(',',expand=True)[1].astype(int)
experimental = experimental[['source','target','epistasis','labels','distance']]

experimental_2 = experimental.copy()
experimental_2.columns = ['target','source','epistasis','labels','distance']

experimental = pd.merge(experimental,innocenti,left_on=['source','target'],right_on=['feature_codon_source','feature_codon_target'],how='left')
experimental_2 = pd.merge(experimental_2,innocenti,left_on=['source','target'],right_on=['feature_codon_source','feature_codon_target'],how='left')
experimental = pd.concat([experimental,experimental_2])

experimental =pd.merge(experimental,epi_data,left_on=['source','target'],right_on=['source_position','position'],how='left')
experimental
```

```python
experimental = experimental[['source','target','epistasis','mi','outlier','masked_change','significant_masked_change','single_significant_masked_change','labels']]
experimental['std_greater'] = abs(experimental.epistasis) > abs(experimental.epistasis).mean()+abs(experimental.epistasis).std()
experimental
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, recall_score,f1_score

def sensitivity_specificity(y_true, y_score):
    y_pred = y_score
    y_true = np.asarray(y_true).astype(int)

    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()

    sensitivity = recall_score(y_true, y_pred)
    specificity = tn / (tn + fp) if (tn + fp) > 0 else np.nan
    f1 = f1_score(y_true, y_pred, pos_label=1)

    return { "sensitivity": sensitivity, "specificity": specificity,'f1':f1}
```

```python
sensitivity_specificity(experimental.mi.isna(),experimental['std_greater'])
```

```python
sensitivity_specificity(experimental.significant_masked_change.isna(),experimental['std_greater'])
```

# Statistical Test for Enrichment

```python
epi_data
```

```python
rbd_mutations = epi_data[(epi_data.mutation.str[1:-1].astype(int) >= 319) & (epi_data.mutation.str[1:-1].astype(int) <= 514) ]
rbd_mutations = rbd_mutations[rbd_mutations.masked_change.isna() == False]
rbd_mutations
```

```python
rbd_positions = rbd_mutations.mutation.str[1:-1].astype(int).unique()
rbd_positions
```

## Enrichment of RBD experimental results in detected interactions

```python
rbd_mutations['in_experimental_set'] = [True if position in rbd_positions else False for position in rbd_mutations.position]    
```

```python
pvals = []
```

```python
population_size = len(rbd_mutations)
n_successes_in_population = len(rbd_mutations[rbd_mutations.in_experimental_set == True])

n_samples = len(rbd_mutations[rbd_mutations.significant_masked_change.isna() == False])
n_successes = len(rbd_mutations[(rbd_mutations.significant_masked_change.isna() == False) & (rbd_mutations.in_experimental_set == True)])
population_size,n_successes_in_population,n_samples,n_successes

from scipy.stats import hypergeom
pval = hypergeom.sf(n_successes - 1, population_size, n_successes_in_population, n_samples)
# Expected number of successes
expected = n_samples * (n_successes_in_population / population_size)
# Enrichment (fold-change)
enrichment = (n_successes / n_samples) / (n_successes_in_population / population_size)

pvals.append(pval)
pval,enrichment
```

## Enrichment of high impact RBD experimental results in detected interactions

```python
def check_significant(source,target,labels):
    if str(source)+','+str(target) in labels or str(target)+','+str(source) in labels:
        return True
    return False 
```

## Enrichment of Spike mutation-mutation results in detected interactions

```python
mutation_positions = epi_data.mutation.str[1:-1].astype(int).unique()
mutation_positions
```

```python
spike_data = epi_data[epi_data.masked_change.isna() == False]
```

```python
spike_data['mutation_site'] = [True if position in mutation_positions else False for position in spike_data.position]    
```

```python
population_size = len(spike_data)
n_successes_in_population = len(spike_data[spike_data.mutation_site == True])

n_samples = len(spike_data[spike_data.significant_masked_change.isna() == False])
n_successes = len(spike_data[(spike_data.significant_masked_change.isna() == False) & (spike_data.mutation_site == True)])
population_size,n_successes_in_population,n_samples,n_successes

from scipy.stats import hypergeom
pval = hypergeom.sf(n_successes - 1, population_size, n_successes_in_population, n_samples)
# Expected number of successes
expected = n_samples * (n_successes_in_population / population_size)
# Enrichment (fold-change)
enrichment = (n_successes / n_samples) / (n_successes_in_population / population_size)

pvals.append(pval)
pval,enrichment
```

```python
pvals
```

```python
from scipy.stats import false_discovery_control
p_adj_bh = false_discovery_control(pvals, method="bh")
significant_indices = np.where(p_adj_bh < 0.05)[0]
significant_indices,p_adj_bh
```
