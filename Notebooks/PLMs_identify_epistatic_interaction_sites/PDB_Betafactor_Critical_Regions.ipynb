```python
import pandas as pd
from Functions import *

representations = decompress_pickle('DMS/SARS-CoV-2_3B_S/S/representations/DMS_S.pbz.pbz2')
```

```python
index = [ k for k in representations['S'].keys()]
values_df = pd.DataFrame([ representations['S'][v]['logits'] for v in representations['S'].keys()],index= index)
values_df
```

```python
values_df_reference = values_df.loc['M1M']
values_df = values_df[values_df.index.str.contains('mask') == False]
values_df = values_df[values_df.index.str[0] != values_df.index.str[-1] ]
values_df
```

```python
change_values_df = values_df - values_df_reference

#Filter values where mutations are at the affected position
filter_mutated_position_change_values_df = pd.DataFrame([ list(row.values[0:int(i[1:-1])-1])+[np.nan]+list(row.values[int(i[1:-1]):]) for i,row in change_values_df.iterrows()])
filter_mutated_position_change_values_df.columns = filter_mutated_position_change_values_df.columns.astype(int)+1
filter_mutated_position_change_values_df.index = change_values_df.index
filter_mutated_position_change_values_df
```

```python

def significant_values(df,significance_type = None):
    # Checks if a change is significant when compared to all changes observed at any position
    if significance_type == 'all':
        df = df.melt(ignore_index=False).reset_index()
        df.columns = ['mutation', 'position', 'change']
        df['abs_change'] =abs(df.change)
        df["significant_change"] = (df.change > df.change.mean() + (df.change.std()*2)) | (df.change < df.change.mean() - (df.change.std()*2))
        df = df[df["significant_change"] == True]
        df = df.dropna(subset=['change','abs_change'])
        df['mutated_position'] = df.mutation.str[1:-1].astype(int)
        df.columns = ['mutation', 'affected_position', 'change', 'abs_change', 'significant_change','mutated_position']

        
    # Checks if a change is significant when compared to all changes observed at the same position
    elif significance_type == 'position':
        df = df[(df < df.mean()-df.std()*2) | (df > df.mean()+df.std()*2) ]
        # df = df.abs().groupby(df.index.str[1:-1].astype(int))
        df = df.melt(ignore_index=False).reset_index()
        df.columns = ['mutation', 'position', 'change']
        df['abs_change'] =abs(df.change)
        df['significant_change'] = True
        df['mutated_position'] = df.mutation.str[1:-1].astype(int)
        df = df.dropna(subset=['change','abs_change'])
        df.columns = ['mutation', 'affected_position', 'change', 'abs_change', 'significant_change','mutated_position']

    # Checks if a change is significant when compared to all changes observed at the same position
    elif significance_type == 'individual':
        df =  df.T
        df = df[(df < df.mean()-df.std()*2) | (df > df.mean()+df.std()*2) ]
        df =  df.T
        # df = df.abs().groupby(df.index.str[1:-1].astype(int))
        df = df.melt(ignore_index=False).reset_index()
        df.columns = ['mutation', 'position', 'change']
        df['abs_change'] =abs(df.change)
        df['significant_change'] = True
        df['mutated_position'] = df.mutation.str[1:-1].astype(int)
        df = df.dropna(subset=['change','abs_change'])
        df.columns = ['mutation', 'affected_position', 'change', 'abs_change', 'significant_change','mutated_position']

    df_mutation_affects_no_sites_count = df.mutated_position.value_counts().reset_index()
    df_mutation_affects_no_sites_count.columns = ['mutated_position','mutation_affects_n_sites']
    df_mutation_affects_no_sites_count.mutation_affects_n_sites = df_mutation_affects_no_sites_count.mutation_affects_n_sites.astype(int)

    df_site_affected_by_mutation_count = df.affected_position.value_counts().reset_index()
    df_site_affected_by_mutation_count.columns= ['affected_position','site_affected_by_n_mutations']
    df_site_affected_by_mutation_count.site_affected_by_n_mutations = df_site_affected_by_mutation_count.site_affected_by_n_mutations.astype(int)

    df = pd.merge(df,df_mutation_affects_no_sites_count,left_on='mutated_position',right_on='mutated_position',how='left')
    df = pd.merge(df,df_site_affected_by_mutation_count,left_on='affected_position',right_on='affected_position',how='left')
    df = df.sort_values(['mutated_position','affected_position'])   

    return df
```

```python
significant_all = significant_values(filter_mutated_position_change_values_df,'all')
# significant_positional = significant_values(filter_mutated_position_change_values_df,'position')
# significant_individual = significant_values(filter_mutated_position_change_values_df,'individual')
```

```python
input_file = '/home/kieran/raid/Github/ESM2_Paper_Revision_Data/PDB/6vsb_1_1_1.pdb'
```

```python
def edit_bfactor(input_file,scores,attribute,is_mutations=False,make_positive=False):
    all_mutation_scores = []
    if is_mutations:
        for mutation in scores.mutation.unique():
            mutation_scores = scores[scores.mutation == mutation]
            all_mutation_scores.append(mutation_scores)
    else:
        all_mutation_scores.append(scores)
    for mutation_scores in all_mutation_scores:
        if is_mutations:
            mutation = mutation_scores.mutation.values[0]
        else:
            mutation = 'reference'

        if 'pos' in mutation_scores.columns:
            position_col = 'pos'
        else:
            position_col = 'position'

        lines = []
        with open(input_file, 'r') as file:
            for l in file:
                if 'ATOM' in l:
                    residue_number = int(l[22:26].split()[0])
                    if len(mutation_scores[mutation_scores[position_col] == residue_number]) == 0:
                        change = 0
                        print('position not present in file, value defaulting to 0')
                    else:
                        change = float(mutation_scores[mutation_scores[position_col] == residue_number][attribute].values[0])
                    if make_positive:
                        change = abs(change)
                    l = l[:60] + str(round(change,2))[:6] +l[66:]
                lines.append(l)

            output_file = input_file.split('.')[0]+'_'+mutation+'_'+attribute+'_'+input_file.split('.')[1]+'.pdb'
            with open(output_file, 'w') as file:    
                for l in lines:
                    file.write(l)
                    file.write('\n')
```
