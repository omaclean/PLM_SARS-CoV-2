```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
fine_tuned=pd.read_csv("/home2/oml4h/PLM_SARS-CoV-2/Results/H3_epistasis_mutation_info_spyros_model.csv")

OG=pd.read_csv("/home2/oml4h/PLM_SARS-CoV-2/Results/Influenza_H3N2_HA_epistasis_mutation_effects_raw_ESM2.csv")

print(OG.columns)

comb=pd.merge(fine_tuned,OG,on=["Mutation","lineage_backbone"],how="inner",suffixes=("_finetuned","_OG"))
```

```python
# create a two panel plot one logged, one not

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)


sns.scatterplot(data=comb,x="probability_finetuned",y="probability_OG",hue="lineage_backbone")
plt.title("Linear Scale")
plt.subplot(1, 2, 2)
# log scale
sns.scatterplot(data=comb,x="probability_finetuned",y="probability_OG",hue="lineage_backbone")
plt.title("Log-Log Scale")
# log scale
plt.xscale("log")
plt.yscale("log")
```
