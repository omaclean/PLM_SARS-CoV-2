```python
import sys
sys.path.append('../../')
import pandas as pd
from scipy.stats import mannwhitneyu
```

```python
dms = pd.read_csv('./DMS/DMS_S_annotated.csv')
```

```python
dms_subunit_comparison = dms[['pos','relative_grammaticality','semantic_score','subunit','Escape','Binding','Entry','Entropy']]
dms_subunit_comparison  = dms_subunit_comparison.groupby('pos').agg({
    'relative_grammaticality':'mean',
    'semantic_score':'mean',
    'Escape':'mean',
    'Binding':'mean',
    'Entry':'mean',
    'Entropy':'first',
    'subunit':'first'
    })
```

```python
dms_s1 = dms_subunit_comparison[dms_subunit_comparison.subunit == 'S1']
dms_s2 = dms_subunit_comparison[dms_subunit_comparison.subunit == 'S2']
```

```python
for feature in ['relative_grammaticality','semantic_score','Escape','Binding','Entry','Entropy']:
    p = mannwhitneyu(
             x=dms_s1[feature].dropna()	,
             y=dms_s2[feature].dropna()	,
             alternative = 'two-sided',
             nan_policy='raise').pvalue
    significant = p<(0.05/6)
    print(feature,p,significant)
```
