```python
import pandas as pd
import numpy as np
```

```python
spike_dms_data = pd.read_csv('./DMS/DMS_S:0.csv')
spike_dms_data
```

```python
reference_spike_dms_data =  spike_dms_data[spike_dms_data.label.str[0] == spike_dms_data.label.str[-1] ]
reference_spike_dms_data
```

```python
mask_spike_dms_data = spike_dms_data[spike_dms_data.label.str.contains('mask')]
mask_spike_dms_data
```

```python
amino_acid_spike_dms_data = spike_dms_data[(spike_dms_data.label.str.contains('mask') == False) & (spike_dms_data.label.str[0] != spike_dms_data.label.str[-1] )] 
amino_acid_spike_dms_data
```

```python
amino_acid_spike_dms_data = amino_acid_spike_dms_data.sort_values('semantic_score')
amino_acid_spike_dms_data['semantic_rank'] = amino_acid_spike_dms_data.reset_index().index.astype(int) + 1
amino_acid_spike_dms_data = amino_acid_spike_dms_data.sort_values('grammaticality')
amino_acid_spike_dms_data['grammatical_rank'] =amino_acid_spike_dms_data .reset_index().index.astype(int) + 1
amino_acid_spike_dms_data['acquisition_priority'] = amino_acid_spike_dms_data['semantic_rank'] + amino_acid_spike_dms_data['grammatical_rank']

amino_acid_spike_dms_data = amino_acid_spike_dms_data.sort_values('sequence_grammaticality')
amino_acid_spike_dms_data['sequence_grammatical_rank'] =amino_acid_spike_dms_data.reset_index().index.astype(int) + 1
amino_acid_spike_dms_data['sequence_acquisition_priority'] = amino_acid_spike_dms_data['semantic_rank'] + amino_acid_spike_dms_data['sequence_grammatical_rank']
```

```python
mask_spike_dms_data['semantic_rank'] = np.nan
mask_spike_dms_data['grammatical_rank'] = np.nan
mask_spike_dms_data['acquisition_priority'] = np.nan
mask_spike_dms_data['sequence_grammatical_rank'] = np.nan
mask_spike_dms_data['sequence_acquisition_priority'] = np.nan

reference_spike_dms_data['semantic_rank'] = np.nan
reference_spike_dms_data['grammatical_rank'] = np.nan
reference_spike_dms_data['acquisition_priority'] = np.nan
reference_spike_dms_data['sequence_grammatical_rank'] = np.nan
reference_spike_dms_data['sequence_acquisition_priority'] = np.nan
```

```python
mask_spike_dms_data
```

```python
average_amino_acid_spike_dms_data = amino_acid_spike_dms_data[['ref','pos','semantic_score', 'relative_grammaticality', 'relative_sequence_grammaticality','relative_mutated_grammaticality','relative_masked_grammaticality','grammaticality','sequence_grammaticality','sequence_grammaticality','mutated_grammaticality','masked_grammaticality']]
average_amino_acid_spike_dms_data = average_amino_acid_spike_dms_data.groupby(['ref','pos']).mean().reset_index()
average_amino_acid_spike_dms_data.columns = ['ref','pos'] + ['mean_'+col for col in average_amino_acid_spike_dms_data.columns[2:]]
average_amino_acid_spike_dms_data
```

```python
reranked_dms = pd.concat([amino_acid_spike_dms_data, mask_spike_dms_data,reference_spike_dms_data]).sort_values(['pos','label'])
reranked_dms
```

```python
reranked_dms = pd.merge(reranked_dms,average_amino_acid_spike_dms_data,how='left',left_on=['ref','pos'], right_on=['ref','pos'] )
reranked_dms
```

```python
wuhan_rbd_dms = rbd_dms[rbd_dms.target == 'Wuhan-Hu-1']
wuhan_rbd_dms.columns = ['target','position','mutant','wuhan_1_binding','wuhan_1_delta_binding','wuhan_1_expression','wuhan_1_delta_expression']
```

```python
compiled_rbd_dms = pd.merge(wuhan_rbd_dms,
                            rbd_variant_dms_avg[['position','mutant','variant_average_binding','variant_average_delta_binding','variant_average_expression','variant_average_delta_expression']],
                            how='left',
                            left_on=['position','mutant'],
                            right_on = ['position','mutant'],
                           )
compiled_rbd_dms = compiled_rbd_dms.drop('target',axis=1)
compiled_rbd_dms
```

```python
all_dms = pd.merge(compliled_full_dms,compiled_rbd_dms.drop(['wuhan_1_delta_binding','wuhan_1_delta_expression','variant_average_delta_binding','variant_average_delta_expression'],axis=1),how='left',
                            left_on=['position','mutant'],
                            right_on = ['position','mutant'],)
all_dms
```

```python
# ESM + Evescape dissimilarity + Evescape Accessibility
temperatures = {"fitness": 1, "surfacc": 1, "exchangability": 2}

def logistic(x):
    return 1 / (1 + np.exp(-x))

def standardization(x):
    """Assumes input is numpy array or pandas series"""
    return (x - x.mean()) / x.std()

from sklearn.impute import SimpleImputer
#Impute missing values for columns used to calculate EVEscape scores
impute_cols = ["Accessibility WCN", "Dissimilarity Charge Hydro"]
df_imp = reranked_dms[impute_cols].copy()
imp = SimpleImputer(missing_values=np.nan, strategy="mean")
df_imp = pd.DataFrame(imp.fit_transform(df_imp),
                        columns=df_imp.columns,
                        index=df_imp.index)
reranked_dms = reranked_dms.drop(["Accessibility WCN", "Dissimilarity Charge Hydro"],axis=1)
reranked_dms = pd.concat([reranked_dms,df_imp],axis=1)

reranked_dms['relative_grammaticality_with_eve']= 0
reranked_dms['relative_grammaticality_with_eve'] += np.log(logistic(
            standardization(reranked_dms["relative_grammaticality"]) * 1 /
            temperatures["fitness"]))

reranked_dms['relative_grammaticality_with_eve']  += np.log(
        logistic(
            standardization(reranked_dms["Accessibility WCN"]) * 1 /
            temperatures["surfacc"]))

reranked_dms['relative_grammaticality_with_eve']  += np.log(
        logistic(
            standardization(reranked_dms["Dissimilarity Charge Hydro"]) * 1 /
            temperatures["exchangability"]))


reranked_dms['ESM-1b Relative Grammaticality with EVE']= 0
reranked_dms['ESM-1b Relative Grammaticality with EVE'] += np.log(logistic(
            standardization(reranked_dms["ESM-1b Relative Grammaticality"]) * 1 /
            temperatures["fitness"]))

reranked_dms['ESM-1b Relative Grammaticality with EVE']  += np.log(
        logistic(
            standardization(reranked_dms["Accessibility WCN"]) * 1 /
            temperatures["surfacc"]))

reranked_dms['ESM-1b Relative Grammaticality with EVE']  += np.log(
        logistic(
            standardization(reranked_dms["Dissimilarity Charge Hydro"]) * 1 /
            temperatures["exchangability"]))
```
